```python
ratings_with_books = ratings.merge(books,on='ISBN')
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'rating':'number of ratings'},inplace=True)
```

```python
number_rating
```

```python
final_rating = ratings_with_books.merge(number_rating,on='title')
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating=final_rating[final_rating['number of ratings']>=50]
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
# Create Pivot Table
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
# convert it into parse matrix
# while calculating eclidean distance no need to add lot of zeros
# to save computation time
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors # clustering algorithm
```

```python
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[54, :].values.reshape(1,-1), n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in suggestions[0]:
    print(book_pivot.index[i])
```

```python
book_pivot.index[54]
```

```python
def test(id):
    distances, suggestions = model.kneighbors(book_pivot.iloc[id, :].values.reshape(1,-1), n_neighbors=6)
    for i in suggestions[0]:
        print(book_pivot.index[i])
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index==book_name)[0][0]
    distances, suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1,-1), n_neighbors=6)
    for i in suggestions[0][1:]:
        print(i,':',book_pivot.index[i])
```

```python
recommend_book('Animal Farm')
```

```python
recommend_book('Exclusive')
```

```python
recommend_book('Harry Potter and the Chamber of Secrets (Book 2)')
```
