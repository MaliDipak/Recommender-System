```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get('https://www.imdb.com/search/title/?release_date=1980-01-01,2022-12-31&colors=color&countries=in&languages=hi&adult=include&count=250',headers=headers).text
```

```python
title = []
director = []
actors = []
year = []
genre = []
rating = []
desc = []
votes = []
runtime = []
imdbid = []

for i in movies:
  try:
    a = i.find_all('a')
    a = a[:1] + a[12:]
    t = a[0].text.strip()
    

    dir = a[1].text.strip()
    
    act = a[2:]
    act = [j.text.strip() for j in act]
    
    y = i.find_all('span',class_='lister-item-year')[0].text.strip()
    
    g = i.find_all('span',class_='genre')[0].text.strip()
    
    rat = i.find_all('div',class_='ratings-imdb-rating')[0].text.strip()
    
    des = i.find_all('p',class_='text-muted')[1].text.strip()

    v = i.find_all('p',class_='sort-num_votes-visible')[0].text.strip()[7:]

    rt = i.find_all('span',class_='runtime')[0].text.strip()


    anr = i.find_all('a',)
    id = anr[0]['href'].split('/')[-2]

    imdbid.append(id)
    runtime.append(rt)
    title.append(t)
    director.append(dir)
    actors.append(act)
    year.append(y)
    genre.append(g)
    rating.append(rat)
    desc.append(des)
    votes.append(v)
  except Exception as e:
    continue

```

```python
# To fetch movies from imdb
num = 1
start = 251
title = []
director = []
actors = []
year = []
genre = []
rating = []
desc = []
votes = []
runtime = []
imdbid = []

while(start<10251):
  try:
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

    url = "https://www.imdb.com/search/title/?release_date=1980-01-01,2022-12-31&colors=color&countries=in&languages=hi&adult=include&count=250&start={}&ref_=adv_nxt".format(start)

    webpage = requests.get(url,headers=headers).text

    soup = BeautifulSoup(webpage,'lxml')

    movies = soup.find_all('div',class_='lister-item-content')

    for i in movies:
      try:
        a = i.find_all('a')
        a = a[:1] + a[12:]
        t = a[0].text.strip()
        

        dir = a[1].text.strip()
        
        act = a[2:]
        act = [x.text.strip() for x in act]
        
        y = i.find_all('span',class_='lister-item-year')[0].text.strip()
        
        g = i.find_all('span',class_='genre')[0].text.strip()
        
        rat = i.find_all('div',class_='ratings-imdb-rating')[0].text.strip()
        
        des = i.find_all('p',class_='text-muted')[1].text.strip()

        v = i.find_all('p',class_='sort-num_votes-visible')[0].text.strip()[7:]

        rt = i.find_all('span',class_='runtime')[0].text.strip()

        anr = i.find_all('a',)
        id = anr[0]['href'].split('/')[-2]


        imdbid.append(id)
        runtime.append(rt)
        title.append(t)
        director.append(dir)
        actors.append(act)
        year.append(y)
        genre.append(g)
        rating.append(rat)
        desc.append(des)
        votes.append(v)

      except Exception as e:
        continue

  except Exception as e:
    print("Error: ",url)


  print("-----------------------------------")
  print("Epoch ",num)
  print(url)
  print(title[-1])
  num = num+1

  start = start + 250



```

```python
next250 = pd.DataFrame({"ImdbId":imdbid,"Title":title,"Year":year,"Director":director,"Actors":actors,"Genre":genre,"Description":desc,"Votes":votes,"Runtime":runtime})
```
